```python
import wbdata
import pandas as pd
import datetime

# 1. Indicators (8 Features + 1 Target)
indicators = {
    'NY.GDP.MKTP.KD.ZG': 'GDP_Growth',
    'FP.CPI.TOTL.ZG': 'Inflation',
    'SL.UEM.TOTL.ZS': 'Unemployment',
    'NE.EXP.GNFS.ZS': 'Exports_percent_GDP',
    'BX.KLT.DINV.WD.GD.ZS': 'FDI_Inflow',
    'NE.GDI.TOTL.ZS': 'Capital_Formation',
    'GC.XPN.TOTL.GD.ZS': 'Govt_Expenditure',
    'SP.POP.GROW': 'Population_Growth',
    'FR.INR.RINR': 'Real_Interest_Rate'
}

# 2. Set range and Country to "all"
data_date = (datetime.datetime(1961, 1, 1), datetime.datetime(2023, 1, 1))

print("⏳ Downloading global data... This will take a few minutes.")

# 3. Fetch data for all countries
df = wbdata.get_dataframe(indicators, country="all", data_date=data_date)

# 4. Cleanup
df.reset_index(inplace=True)
df = df.rename(columns={'date': 'Year', 'country': 'Country'})
df['Year'] = df['Year'].astype(int)

# Sort by Country and then Year (Vital for time-series)
df.sort_values(['Country', 'Year'], inplace=True)

# 5. Smart Interpolation (Grouped by Country)
# We use groupby so interpolation happens ONLY within each country's own timeline
df = df.groupby('Country', group_keys=False).apply(
    lambda group: group.interpolate(method='linear', limit_direction='both')
)

# 6. Save
df.to_csv("global_gdp_dataset.csv", index=False)

print(f"✅ Success! Data for {df['Country'].nunique()} entities saved.")
df
```

```python
import wbdata
import pandas as pd
import datetime
import time

# ... (Keep your indicators and data_date as they were) ...

# Get a list of all actual countries (excluding aggregates to keep it small)
countries = [c['id'] for c in wbdata.get_country() if c['region']['value'] != 'Aggregates']

all_data = []

print(f"⏳ Downloading data in chunks for {len(countries)} countries...")

# Fetch data in smaller batches to prevent JSONDecodeError
for i in range(0, len(countries), 50):  # Fetch 50 countries at a time
    chunk = countries[i:i+50]
    try:
        df_chunk = wbdata.get_dataframe(indicators, country=chunk, data_date=data_date)
        all_data.append(df_chunk)
        print(f"✅ Processed {i+len(chunk)} countries...")
        time.sleep(1) # Small delay to be nice to the server
    except Exception as e:
        print(f"❌ Error at chunk starting with {chunk[0]}: {e}")

# Combine everything back together
df = pd.concat(all_data)
df.reset_index(inplace=True)
```

```python
gdp.columns = gdp.columns.str.strip()
gdp['country'] = gdp['country'].str.strip()

# 2. Calculate GDP Growth (Required for the heatmap to make sense)
# (Raw GDP doesn't work well in heatmaps because US/China dwarf everyone else)
gdp = gdp.sort_values(['country', 'Year'])
gdp['GDP_Growth'] = gdp.groupby('country')['GDP_Growth'].pct_change() * 100

# 3. Pivot Data (Transform to Matrix Format)
# Rows = Countries, Columns = Years, Values = Growth %
heatmap_data = gdp.pivot(index='country', columns='Year', values='GDP_Growth')

# 4. Create Heatmap
plt.figure(figsize=(35,45)) # Make it TALL to fit 200 countries

sns.heatmap(
    heatmap_data, 
    cmap='RdYlGn',   # Red = Recession, Green = Growth
    center=0,        # White/Yellow = 0% Growth
    vmin=-10, vmax=10, # Cap colors at -10% and +10% to handle outliers
    cbar_kws={'label': 'GDP Growth (%)'}
)

plt.title("Global Economic Growth Heatmap (1970-2021)")
plt.xlabel("Year")
plt.ylabel("Country")
plt.legend()

plt.show()
```
